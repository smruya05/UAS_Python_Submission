```
# Password validation using loops and strings
correct_pw = "python123"
attempts = 3

while attempts > 0:
    guess = input(f"Enter password ({attempts} attempts left): ")
    if guess == correct_pw:
        print("Access Granted!")
        break
    else:
        attempts -= 1
        print("Wrong password.")

if attempts == 0:
    print("Account Locked.")
```
