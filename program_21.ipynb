```
# Defining a function to calculate total cost with tax
def calculate_total(price, tax_rate=0.05):
    """Calculates total price including a default tax of 5%"""
    total = price + (price * tax_rate)
    return total

item_price = float(input("Enter item price: "))
final_bill = calculate_total(item_price)
print(f"The final bill (including 5% tax) is: ${final_bill:.2f}")
```
