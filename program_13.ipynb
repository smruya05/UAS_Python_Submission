```
# Managing a list of students
students = ["Alice", "Bob", "Charlie"]
print(f"Initial list: {students}")

# Adding a new student
new_student = input("Enter student name to add: ")
students.append(new_student)

# Sorting the list
students.sort()
print(f"Sorted student list: {students}")
print(f"Total students enrolled: {len(students)}")
```
