```
# Using Lambda functions for quick operations
numbers = [1, 2, 3, 4, 5]
square = lambda x: x ** 2

squared_numbers = list(map(square, numbers))
print(f"Original List: {numbers}")
print(f"Squared List: {squared_numbers}")
```
