```
# Check if a number is even or odd using the modulo operator
num_to_check = int(input("Please enter an integer: "))

if num_to_check % 2 == 0:
    print(f"The number {num_to_check} is EVEN.")
else:
    print(f"The number {num_to_check} is ODD.")
print("Concept: A number is even if it leaves no remainder when divided by 2.")
```
