```
# Using Sets to handle unique values
raw_data = [1, 2, 2, 3, 4, 4, 4, 5]
unique_set = set(raw_data)

print(f"Original list with duplicates: {raw_data}")
print(f"Unique set: {unique_set}")

unique_set.add(6)
print(f"Set after adding 6: {unique_set}")
```
