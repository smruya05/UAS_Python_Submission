```
# String manipulation and verbose formatting
sentence = "Learning Python is essential for Data Science"

print(f"Original String: '{sentence}'")
print(f"Total length of string: {len(sentence)} characters")
print(f"Uppercase version: {sentence.upper()}")
print(f"Lowercase version: {sentence.lower()}")
print(f"First word: {sentence[:8]}")
print(f"String in reverse order: {sentence[::-1]}")
print(f"Checking for 'Python': {'Python' in sentence}")
```
