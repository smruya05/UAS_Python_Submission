```
# Program to find the largest value among three inputs
n1 = float(input("First number: "))
n2 = float(input("Second number: "))
n3 = float(input("Third number: "))

if n1 >= n2 and n1 >= n3:
    largest = n1
elif n2 >= n1 and n2 >= n3:
    largest = n2
else:
    largest = n3

print(f"\nThe maximum value found among {n1}, {n2}, and {n3} is {largest}.")
```
