```python
# Using a while loop to calculate cumulative sum
n = int(input("Enter a positive integer: "))
sum_total = 0
counter = 1

while counter <= n:
    sum_total += counter
    counter += 1

print(f"The sum of first {n} natural numbers is: {sum_total}")
```
