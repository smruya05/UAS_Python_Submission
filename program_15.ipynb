```
# Using dictionaries for structured data
employee = {
    "id": 101,
    "name": "John Doe",
    "department": "IT",
    "salary": 55000
}

print(f"Employee Name: {employee['name']}")
employee['salary'] += 5000  # Giving a raise
print(f"Updated Salary: {employee['salary']}")
print(f"Keys in dictionary: {list(employee.keys())}")
```
