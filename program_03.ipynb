```
import math
# Program to calculate area and circumference
radius = float(input("Enter the radius of the circle in cm: "))

area = math.pi * (radius ** 2)
circumference = 2 * math.pi * radius

print(f"\nFor a circle with radius {radius} cm:")
print(f"The Area is: {area:.2f} sq cm")
print(f"The Circumference is: {circumference:.2f} cm")
```
