```
# Verbose demonstration of arithmetic operators
val1 = float(input("Enter the first numerical value: "))
val2 = float(input("Enter the second numerical value: "))

print("\n--- Calculation Results ---")
print(f"Addition ({val1} + {val2}): {val1 + val2}")
print(f"Subtraction ({val1} - {val2}): {val1 - val2}")
print(f"Multiplication ({val1} * {val2}): {val1 * val2}")
if val2 != 0:
    print(f"Division ({val1} / {val2}): {val1 / val2}")
    print(f"Modulus (Remainder): {val1 % val2}")
else:
    print("Division/Modulus: Error (Cannot divide by zero)")
print(f"Exponentiation ({val1} to power {val2}): {val1 ** val2}")
```
