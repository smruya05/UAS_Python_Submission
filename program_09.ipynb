```
# String matching simulator for traffic lights
signal = input("Current Signal Color (Red/Yellow/Green): ").strip().capitalize()

if signal == "Red":
    print("Action: Stop your vehicle and wait.")
elif signal == "Yellow":
    print("Action: Slow down and prepare to stop.")
elif signal == "Green":
    print("Action: Proceed with your journey.")
else:
    print("Error: Unknown signal color encountered.")
```
