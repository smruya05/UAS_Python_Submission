```
# Celsius to Fahrenheit conversion with conditional feedback
temp_c = float(input("Enter the temperature in Celsius: "))
temp_f = (temp_c * 9/5) + 32

print(f"\n{temp_c}°C is equivalent to {temp_f}°F")

if temp_f >= 100:
    print("Warning: Extremely high temperature detected!")
elif temp_f <= 32:
    print("Warning: Freezing temperature detected!")
else:
    print("Temperature is within the moderate range.")
```
