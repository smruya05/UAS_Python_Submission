```
# Grading system demonstration using an if-elif-else ladder
marks = float(input("Enter the student's total marks (0-100): "))

if marks >= 90:
    result = "Grade A+"
elif marks >= 80:
    result = "Grade A"
elif marks >= 70:
    result = "Grade B"
elif marks >= 60:
    result = "Grade C"
elif marks >= 40:
    result = "Grade D"
else:
    result = "Grade F (Fail)"

print(f"Final Assessment Result: {result}")
```
